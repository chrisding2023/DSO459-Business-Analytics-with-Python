```python
# data cleaning for liquidity holders datasets
dai_eth.loc[:,"BALANCE"] = dai_eth.loc[:,"BALANCE"].astype(float) / np.power(10, dai_eth.loc[:,"CONTRACT_DECIMALS"])
dai_eth.loc[:,"TOTAL_SUPPLY"] = dai_eth.loc[:,"TOTAL_SUPPLY"].astype(float) / np.power(10, dai_eth.loc[:,"CONTRACT_DECIMALS"])
uni_eth.loc[:,"BALANCE"] = uni_eth.loc[:,"BALANCE"].astype(float) / np.power(10, uni_eth.loc[:,"CONTRACT_DECIMALS"])
uni_eth.loc[:,"TOTAL_SUPPLY"] = uni_eth.loc[:,"TOTAL_SUPPLY"].astype(float) / np.power(10, uni_eth.loc[:,"CONTRACT_DECIMALS"])
usdc_eth.loc[:,"BALANCE"] = usdc_eth.loc[:,"BALANCE"].astype(float) / np.power(10, usdc_eth.loc[:,"CONTRACT_DECIMALS"])
usdc_eth.loc[:,"TOTAL_SUPPLY"] = usdc_eth.loc[:,"TOTAL_SUPPLY"].astype(float) / np.power(10, usdc_eth.loc[:,"CONTRACT_DECIMALS"])
wbtc_eth.loc[:,"BALANCE"] = wbtc_eth.loc[:,"BALANCE"].astype(float) / np.power(10, wbtc_eth.loc[:,"CONTRACT_DECIMALS"])
wbtc_eth.loc[:,"TOTAL_SUPPLY"] = wbtc_eth.loc[:,"TOTAL_SUPPLY"].astype(float) / np.power(10, wbtc_eth.loc[:,"CONTRACT_DECIMALS"])
```

<h2 style="color:red">For every token holder pool (there are five), calculate the percentage of wallet addresses which have a positive balance. Create a new data frame summarizing these percentages and create a plot that best displays this information.</h2>


```python
pools = [dai_eth, eth_usdt, uni_eth, usdc_eth, wbtc_eth]
```

```python
# positive function:extract the datafram with positive balance
def positive(data):
    data_positive = data.loc[data["BALANCE"]>0]
    return data_positive
# per function to get the percentage of positive balance in each different holder dataset
def per(data,oridata):
    perofdata = data.shape[0]/oridata.shape[0]
    return perofdata
```
